# Example 1
About as simple as they come

```julia
using Random

Random.seed!(42)

N = 100
random_x = range(0, stop=1, length=N)
random_y0 = randn(N) .+ 5
random_y1 = randn(N)
random_y2 = randn(N) .- 5

p1 = plot([
    scatter(x=random_x, y=random_y0, mode="markers", name="markers"),
    scatter(x=random_x, y=random_y1, mode="lines", name="lines"),
    scatter(x=random_x, y=random_y2, mode="markers+lines", name="markers+lines")
])
```

# Example 2
A bit more complicated

```julia
θ = collect(range(0,10,101));
```

```julia
f = sin.(θ);
```

```julia
s1 = scatter(x=θ, y=f, name="sin", mode="lines+markers", line_color="Crimson", line_width=1);
s2 = scatter(x=θ, y=2*f, mode="markers", name="sin2A", marker=attr(symbol="square", color="cyan",
            line=attr(color="navy", width=1)));
data = [s1, s2];
```

```julia
layout = Layout(title="Two Sine Plots",
                xaxis_title="θ", yaxis_title="Sin(θ)",
                xaxis_range=[2, 8], yaxis_range=[-3, 3],
                modebar_add=["drawClosedPath"]);
```

```julia
p2 = plot(data, layout)
```

# Save Example 1 as a standalone HTML file

```julia
open("./example1.html", "w") do io
    PlotlyBase.to_html(io, p1.plot)
end
```
